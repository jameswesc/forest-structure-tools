```python
from pathlib import Path

import geopandas as gpd
import pandas as pd
import numpy as np
import pdal

import xarray as xr
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
data_dir = Path("../data")
plots_lidar_dir = data_dir / "outputs" / "plots" / "lidar"
```

```python
def read_plot(plot_id: str):
    file_path = plots_lidar_dir / f"{plot_id}.copc.laz"
    pl = pdal.Reader(str(file_path)).pipeline()
    pl.execute()
    return pl.arrays[0]
```

```python
points_np = read_plot("AGG_O_01_P1")
```

```python
points_xr = xr.Dataset(
    data_vars={
        "x": ("point_idx", points_np["X"]),
        "y": ("point_idx", points_np["Y"]),
        "y": ("point_idx", points_np["Y"]),
        "rn": ("point_idx", points_np["ReturnNumber"]),
        "nor": ("point_idx", points_np["NumberOfReturns"]),
        "weights": ("point_idx", points_np["Weight"])
    },
    coords={
        "point_idx": np.arange(len(points_np))
    }
)
points_xr.head()
```
