```python
import pdal
import pandas as pd
import numpy as np
import pyvista as pv

pv.set_jupyter_backend('client')
```

```python
def read_point_cloud(file_path):
    pipeline = pdal.Reader(file_path).pipeline()
    pipeline.execute()
    return pd.DataFrame(pipeline.arrays[0])
```

```python
lidar_df = read_point_cloud('../data/outputs/sites/lidar/ULO_237.copc.laz')
lidar_df.head()
```

```python
X = lidar_df["X"]
Y = lidar_df["Y"]
Z = lidar_df["HeightAboveGround"].clip(lower=0)
points = np.array([X, Y, Z]).T
```

```python
mesh = pv.PolyData(points)
mesh.plot(point_size=1, style='points')
```
