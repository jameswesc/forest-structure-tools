```python
from pathlib import Path

import geopandas as gpd
import pandas as pd
import numpy as np
import pdal

import xarray as xr
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

from forest_structure_tools.metrics import forest_structure_metrics
```

```python
data_dir = Path("../data")
plots_lidar_dir = data_dir / "outputs" / "plots" / "lidar"

def read_plot(plot_id: str):
    file_path = plots_lidar_dir / f"{plot_id}.copc.laz"
    pl = pdal.Reader(str(file_path)).pipeline()
    pl.execute()
    return pl.arrays[0]

points_np = read_plot("AGG_O_01_P1")
```

```python
x = points_np['X']
y = points_np['Y']
z = points_np['Z']
nor = points_np['NumberOfReturns']
weights = 1 / nor

```
